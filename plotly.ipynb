```python
import plotly.express as px
import plotly.graph_objects as go
import pandas as pd
```

```python
dfe = pd.read_excel('data/esempio-dati.xlsx') 
```

# Disegnare istorgramma e boxplot 

```python
fig = px.histogram(
        dfe, #dataframe
        x="height", #colonna da usare nel dataframe
        title="Altezze", 
        template='seaborn',
        nbins=20,
        range_x=[110,210],
        opacity=0.5,
        marginal='box' ,
        labels={'height':'Altezze in cm'},
        width=600,
        height=600
    )
fig.show()
```

Sactter plot di peso e altezze con retta di regressione lineare semplice (minimi quadrati `trendline=ols` Ordinari Least Squares`)

```python
fig = px.scatter(dfe,x='weight',y='height',trendline="ols")
fig.show()
#results = px.get_trendline_results(fig)
#results.px_fit_results.iloc[0].summary()
```

## Usando `plotly.graph_objects`

```python
fig = go.Figure(
        data=[
            go.Histogram(
                x=dfe.height,
                xbins=dict(
                      start=100,
                      end=220,
                      size=2), 
                autobinx=False),
                
            ],
        )
fig.update_layout(template='seaborn')
fig.show()
```
